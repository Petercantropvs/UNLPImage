```python
# C√≥digo de ejemplo dado en la clase del 3/5
# ventana nuevo perfil, 2 columnas

import PySimpleGUI as sg

H1_SIZE = 50 #30
TEXT_SIZE = 20 #16
sg.theme('Light Grey 1')
# 30,1 no son pixeles, son tama√±o de caracteres
sg.set_options(element_size=(30,1), font=('Marker Felt', TEXT_SIZE))#('Helvetica', 12)) # esta piola para definir cosas globales

def layout():
    header = [[
        sg.Text('Nuevo Perfil', font=('Marker Felt', H1_SIZE), justification='left'),
        sg.Column([[sg.Button('‚á¶')]], expand_x=True, element_justification='right')
    ]]
    column_1 = [
        [sg.Text('Formulario 1')],
        [sg.Text('Nombre')],
        [sg.Input(size=(35,1))],
        [sg.Text('Apellido')],
        [sg.Input(size=(35,1))],
        [sg.Text('Edad')],
        [sg.Input(size=(35,1))],
        [sg.Button('Enviar')],
    ]
    column_2 = [[sg.Image(source='C:/Users/Usuario/Documents/Pedro/Origami/PNGs\perro.png',
                         key='-AVATAR-',
                         size=(280,280),
                         subsample=3, # Al tama√±o de la imagen original lo reduce en 3
                         pad=((125,125), (0,0)),
                         )
                ],
               [sg.FileBrowse('Seleccionar avatar üëæ',
                             key='-BROWSE-',
                             enable_events=True,
                             change_submits=True, # Para que al terminar emita
                             size=(20,2))
               ]]
    
    grid = [
        [
            sg.Column(column_1),
            sg.Column(column_2, element_justification='center',size=(450,400)),
        ],
    ]
    
    actions = [[sg.Button('üíæ Guardar')]]
    
    return [
        [sg.Column(layout=header, expand_x=True)],
        [sg.Column(layout=grid, justification='center', key='main-column', expand_y = True,
                   pad = ((50,50),(50,50)), scrollable = True, vertical_scroll_only = True)],
        [sg.Column(layout=actions, expand_x=True, element_justification='right')], # el scrollable queda medio feo. Sirve si tengo columnas muy largas
    ]                           # pad nos deja un poquito de espacio a los cotados

def run():
    window = sg.Window('Perfil', layout(), finalize = True, resizable = True)
    window.set_min_size((600,400))#(800,600))#(1024,768)) otros tama√±os t√≠picos
#     window['main-column'].vsb.pack_forget() # Esta l√≠nea oculta la barra para hacer el scroll
    
    while True:
        event, values = window.read()
#         sg.Print(event, values)
        print(event, values)
        if event == sg.WIN_CLOSED or event == 'Exit':
            break
        elif event == 'Enviar':
            print(f'Valores del formulario: #{values}')
        elif event == '-BROWSE-':
            filename = values['-BROWSE-']
            window['-AVATAR-'].update(source=filename, 
                                      size=(280,280), # Hay que repetir estos par√°metros para que al actualizar
                                      subsample=3)    # los vuelva a aplicar


if __name__ == '__main__':
    run()    # Esto s√≥lo se va a llamar cuando se llame a este archivo directamente y no desde otro .py
    
    
# EN UNLPIMAGE (otro .py)
# from unlpimage.windows import profile
# if __name__ == '__main__':
#   profile.run()
#

# Frames hace el redondeado y las columnas
```
